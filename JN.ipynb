```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
dt = 0.01
```

```python
G = 6.67408e-11 # gravitational constant

massEarth = 5.972e24 # mass of the Earth
massObject = 10 # mass of the object
```

```python
y = np.zeros(N+1) # position of the object on the y-axis
v = np.zeros(N) # velocity of the object
a = np.zeros(N-1) # acceleration of the object

t = np.arange(0, (N+0.5)*dt, dt) # time variable

a[0] = 9.81
y[[0,1]] = 150
v[1] = v[0] + a[0]*dt
```

```python
r = np.ones(N+1) + y

for i in range(1, N-1):
    a[i] = -1*G*massEarth/r[i]**2
    
for j in range(1, N-1):
    y[j+1] = y[j] + v[j]*dt + (a[j]*(dt**2)*0.5)
    v[j+1] = v[j] + a[j]*dt
```

```python
plt.title("Graph")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")

plt.plot(t, y, ".")

plt.grid(True)

plt.show()
```
